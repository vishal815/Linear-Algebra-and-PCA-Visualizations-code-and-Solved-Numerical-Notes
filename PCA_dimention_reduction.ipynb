```python
import numpy as np
import matplotlib.pyplot as plt

# Given 2D data points
data = np.array([[2,4],[1,3],[0,1],[-1,0.5]])

#step 1: Mean centering
mean = np.mean(data, axis=0)
centered_data = data - mean

#step 2: calculate the covariance matrix
cov_matrix = np.cov(centered_data, rowvar = False)

#step 3: calculate Eigenvalues and Eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

#sort eigenvalues and eigenvectors in descenging order
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvalues = eigenvectors[:, sorted_indices]

#step 4: Choose Principal Compentent (First Eigenvector)
principal_component = eigenvectors[:, 0]

#step 5: Project Data onto Principal compontent
projected_data = np.dot(centered_data, principal_component)

# Plotting the results
plt.figure(figsize=(8, 6))
plt.scatter(centered_data[:, 0], centered_data[:, 1], label="Original Data")
plt.plot([0, principal_component[0]],[0, principal_component[1]], color = 'red', label = "principal Component")
plt.scatter(projected_data, np.zeros_like(projected_data), color = 'green', label = "Projected Data")
plt.xlabel('Dimension 1')
plt.xlabel('Dimension 2')
plt.title('PCA on 2D Data points')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Given 2D Data Points
data = np.array([[2, 4], [1, 3], [0, 1], [-1, 0.5]])

# Calculate Mean
mean = np.mean(data, axis=0)

# Calculate Covariance Matrix
cov_matrix = np.cov(data, rowvar=False)

# Calculate Eigenvalues and Eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
sorted_indices = np.argsort(eigenvalues)[::-1]
sorted_eigenvalues = eigenvalues[sorted_indices]
sorted_eigenvectors = eigenvectors[:, sorted_indices]

# Choose Principal Component
principal_component = sorted_eigenvectors[:, 0]

# Project Data onto Principal Component
projected_data = np.dot(data - mean, principal_component)

# Visualize Results in 2D
plt.scatter(projected_data, np.zeros_like(projected_data), label='PCA Projection')
plt.scatter(data[:, 0], data[:, 1], color='red', label='Original Data')
plt.xlabel('Principal Component')
plt.ylabel('Zero')
plt.title('PCA Projection vs. Original Data')
plt.legend()
plt.show()

# Visualize Results in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data[:, 0], data[:, 1], np.zeros(data.shape[0]), color='red', label='Original Data')
ax.scatter(projected_data, np.zeros_like(projected_data), np.zeros_like(projected_data), label='PCA Projection')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('PCA Projection vs. Original Data (3D)')
ax.legend()
plt.show()
```
